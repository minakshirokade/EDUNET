```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

#creating a dataframe for pair plot

data={
    "Solar": [1200, 1500, 1300],
    "Wind": [3400, 3600, 3200],
    "Hydropower": [2900, 3100, 2800],
    "Biomass": [2500,2700,2400]
}
df = pd.DataFrame(data)
#Create a pair plot
sns.pairplot(df)
plt.show()

```

```python
#heatmap: showing the colrealtion between different type of the Energy

#creating a heatmap to show the corelation matrix
correlation_matrix = df.corr()

#plot heatmap
sns.heatmap(correlation_matrix, annot=True , cmap='coolwarm', linewidths=0.5)#annotation, cmap=colormap

#Add title
plt.title('Corrrelation between Energy sources')
plt.show()

#there is possiblity of having negative correlation values
# In correlation values look for the highest value closest to 1 and -1
```

```python
#BoxPlot: Comparing
#between min and max values it is known as outliers
#upper quartile is the 25% of next to median
#lower quartile is the 25% of next to median

#create a box plot
sns.boxplot(data=df)

#add title
plt.title('Distribution of Energy Consumption by Source')
plt.show()

```

```python
#IRIS FLOWER DATASET : VERY VAMOUS DATASET

#Violin Plot: showing the distribution of the data
#create a violin in plot
sns.violinplot(data=df)

#Add title
plt.title('Violin PLot of Energy Consumption Distribution')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

#Sample data for energy consumption and carbon emissions
energy_values = [100,200,300,400,500]
carbon_emissions= [10,20,25,40,50]

#create a dataframe with energy consumption ad carbon emissions
df_reg = pd.DataFrame({'Energy Consumption': energy_values,
                      'Carbon Emissions': carbon_emissions})

#Create a regression plot
sns.regplot(x='Energy Consumption', y='Carbon Emissions', data=df_reg)

#Add title
plt.title('Regression Plot of Energy Consumption vs Carbon Emissions')
plt.show()
```

```python
#Plotting 3D lines and Plots
#importing m
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()

#syntax for 3-D projection
ax = plt.axes(projection='3d')

#defining all 3 axes
z = np.linspace(0,1,100)
x  = z*np.sin(25*z)
y =z*np.cos(25*z)

#plotting
ax.plot3D(x,y,z, 'green')
ax.set_title('3D Line Plot')
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame(data)
#Create a pair plot
sns.pairplot(df)
plt.show()
```

```python
correlation_matrix= data.iloc[:, :-1].corr()


#plot heatmap
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm',linewidths=0.5)

#add title
plt.title('Heatmap of Iris Feature Correlations')

plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

# Assign numerical values to the species column
label_encoder = LabelEncoder()
df['Species'] = label_encoder.fit_transform(df['species'])

# Filter numeric columns
numeric_df = df.select_dtypes(include='number')

# Calculate the correlation matrix
correlation_matrix = numeric_df.corr()

# Plot the heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)

# Set the title
plt.title('Correlation between iris-species')

# Show the plot
plt.show()
```
